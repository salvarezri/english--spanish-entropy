```python
def colapse_accentuation(dic):
  if "á" in dic:
    dic["a"] += dic["á"]
    del dic["á"]
  if "é" in dic:
    dic["e"] += dic["é"]
    del dic["é"]
  if "í" in dic:
    dic["i"] += dic["í"]
    del dic["í"]
  if "ó" in dic:
    dic["o"] += dic["ó"]
    del dic["ó"]
  if "ú" in dic:
    dic["u"] += dic["ú"]
    del dic["ú"]
  return dic
```

```python
def letter_frecuency(lines, lan):
  letter_freq = {}
  # select letters to count
  if lan == "es":
    alphabet = "abcdefghijklmnñopqrstuvwxyzáéíóú"
  elif lan == "en":
    alphabet = "abcdefghijklmnopqrstuvwxyzñ"
  # Iterate over all lines and characters of the text
  for line in lines:
    for letter in line:
      # cast to lowercase
      letter = letter.lower()
      # check if letter is in alphabet
      if not(letter in alphabet):
        continue
      # count letter
      if letter in letter_freq:
        letter_freq[letter] += 1
      else:
        letter_freq[letter] = 1
        
  # colapse accentuation (ex: á -> a, é -> e, etc.)
  letter_freq = colapse_accentuation(letter_freq)
  return letter_freq
```

```python

def calc_letter_rate(letter_freq):
  d = letter_freq
  # sort the dictionary by the number of occurrences
  sorted_d = sorted(d.items(), key=lambda x: x[1], reverse=True)
  letter_rate = {}
  letter_count = 0
  # calculate the total number of letters
  for letter in sorted_d:
    letter_count += letter[1]
  # calculate the rate of each letter
  for letter in sorted_d:
    letter_rate[letter[0]] = letter[1] / letter_count
  return letter_rate
```

```python
def empty_dictionary(alphabet):
  d = {}
  for letter in alphabet:
    d[letter] = 0
  return d
```

```python
def letter_rate(files_list, lan):
  # create an empty dictionary with the alphabet
  dictionary = empty_dictionary("abcdefghijklmnopqrstuvwxyzñ")
  # iterate over all files
  for file_path in files_list:
    # open the file
    file =  open(file_path, "r", encoding="utf8")
    lines = []
    # read all lines
    for line in file:
      lines.append(line)
    file.close()
    #obtain the letter frequency of the book
    book_dictionary = letter_frecuency(lines, lan)
    for key in book_dictionary:
      # update the letter count of this book
      dictionary[key] += book_dictionary[key]
  print("----"*20)
  print("Common usage frecuency ",lan)
  for key in dictionary:
    print(key," | ", dictionary[key])
  # calculate the letter rate
  letter_rate = calc_letter_rate(dictionary)
  return letter_rate
    
    
```

```python
# list all files on a folder
import os
def list_files(directory):
  files = []
  for file in os.listdir(directory):
    if file.endswith(".txt"):
      files.append(directory + "/" + file)
  return files
```

```python
# load books and calculate letter rate (common use)
files_list_en = list_files("./english")
files_list_es = list_files("./espanol")
txt_en_rate = letter_rate(files_list_en, "en")
txt_es_rate = letter_rate(files_list_es, "es")
print("en: ", txt_en_rate)
print("es: ", txt_es_rate)
```

Ahora calcularemos la entropía usando la siguiente formula
$$
Entropy (H) = -\sum_{i=1}^{n} p(x_i) \log_{2}(p(x_i))
$$

```python
# calculate entropy
import math
def calc_entropy(letter_rate = {"a":0.1}):
  entropy = 0
  for letter in letter_rate.keys():
    entropy += letter_rate[letter] * math.log2(letter_rate[letter])
  entropy = entropy*-1
  return entropy
```

```python
dic_en_entropy = calc_entropy(dic_en_rate)
dic_es_entropy = calc_entropy(dic_es_rate)
txt_en_entropy = calc_entropy(txt_en_rate)
txt_es_entropy = calc_entropy(txt_es_rate)
print("en: ", dic_en_entropy, txt_en_entropy)
print("es: ", dic_es_entropy, txt_es_entropy)
```
